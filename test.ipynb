```python
import pandas as pd

data = [['Dog', 'Golden Retriever', 1, 5], ['Dog', 'German Shepherd', 2, 5], ['Dog', 'Mule', 200, 1], ['Cat', 'Shirazi', 5, 2], ['Cat', 'Siamese', 3, 3], ['Cat', 'Sphynx', 7, 4]]
queries = pd.DataFrame(data, columns=['query_name', 'result', 'position', 'rating']).astype({'query_name':'object', 'result':'object', 'position':'Int64', 'rating':'Int64'})
```

```python
df_new = pd.DataFrame({
    'query_name': ['lfdxfi', 'meayln', 'phqghu', 'rcvscx', 'lfdxfi'],
    'result': ['qduxwfnfozvsr', 'prepggxrpnrvy', 'wcysyycqpevik', 'mznimkkasvwsr', 'kycxfxtlsgyps'],
    'position': [2, 1, 1, 1, 2],
    'rating': [5, 1, 2, 4, 2]
})
```

```python
df_new["quality"] = df_new["rating"] / df_new["position"]
```

```python
queries
```

```python
df_new["poor_query_percentage"] = df_new.groupby("query_name")["rating"].transform(lambda s: (s < 3).mean() * 100)
```

```python
df_new
```

```python
df = df_new.groupby(
    by=["query_name", "poor_query_percentage"]
).agg(
    {"quality" : "mean"}
).reset_index()
```

```python
df["quality"] = df["quality"].round(2)
df["poor_query_percentage"] = df["poor_query_percentage"].round(2)
```
